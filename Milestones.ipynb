# Milestones In Python History #

Just a list of some of the more notable features of various Python releases.

* [3.10](https://docs.python.org/3/whatsnew/3.10.html), 2021 Oct 9
 * Structural pattern matching (`match`-statement)
 * New type-annotation features
 * `distutils` module is deprecated
 * Parenthesized context managers

* [3.9](https://docs.python.org/3/whatsnew/3.9.html), 2020 Oct 9
 * “|” and “|=” operators for `dict`
 * relaxed grammar restrictions on decorators
 * IANA Time Zone Database included in `zoneinfo` module
 * new `str.removeprefix()` and `str.removesuffix()` methods
 * new PEG-based parser for CPython
 * Python 2 backward-compatibility features now trigger deprecation warnings

* [3.8](https://docs.python.org/3/whatsnew/3.8.html), 2019 Oct 14
 * “`:=`” operator
 * positional-only function arguments
 * same ABI for debug and release builds
 * “=” option in formatted string literals
 * Auditing hooks, as per [PEP578](https://www.python.org/dev/peps/pep-0578/).
 * new C API for configuring Python initialization
 * provisional “vectorcall” protocol in C API

* [3.7](https://docs.python.org/3/whatsnew/3.7.html), 2018 Jun 27
 * `async` and `await` are now reserved words
 * new nanosecond-resolution functions in `time` module
 * more `asyncio` module enhancements
 * forced UTF-8 runtime mode
 * customization of access to module attributes
 * new library modules: `contextvars`, `dataclasses`, `importlib.resources`

* [3.6](https://docs.python.org/3/whatsnew/3.6.html), 2016 Dec 23
 * formatted string literals
 * syntax for variable annotations
 * underscores in numeric literals
 * asynchronous generators
 * asynchronous comprehensions
 * `__init__subclass__` to customize subclass creation without needing a metaclasss
 * `__set_name__` in descriptors
 * `os.PathLike`
 * local time disambiguation
 * special methods can be set to `None`
 * more `asyncio` module enhancements
 * propagation of `StopIteration` from within generator now always prohibited

All releases below here are obsolete, according to [this table of versions](https://en.wikipedia.org/wiki/History_of_Python#Table_of_versions).

* [3.5](https://docs.python.org/3/whatsnew/3.5.html), 2015 Sep 13, support ended 2020 Sep 30.
 * coroutines with `async`/`await` syntax
 * “%”-style formatting for `bytes` and `bytearray` types
 * new “@” infix operator and corresponding special method names
 * provisional `typing` module
 * automatic retry of system calls on `EINTR`
 * deprecate propagation of `StopIteration` from within a generator; prohibit entirely with `from __future__ import generator_stop`
 * add `isclose()` functions to `math` and `cmath` modules
 * no more `.pyo` files
 * `asyncio` module enhancements

* [3.4](https://docs.python.org/3/whatsnew/3.4.html), 2014 Mar 16, support ended 2019 Mar 18.
 * newly-created file descriptors are non-inheritable
 * first version of `asyncio` module
 * `enum` module
 * `selectors` module

* [3.3](https://docs.python.org/3/whatsnew/3.3.html), 2012 Sep 29, support ended 2017 Sep 29.
 * `yield from`
 * virtual environments
 * implicit namespace packages (no more need for `__init__.py` files)
 * `raise` ... `from None`
 * bring back `u`-prefix for (redundantly) indicating Unicode strings
 * `__qualname__` attribute
 * `sys.implementation`

* [3.2](https://docs.python.org/3/whatsnew/3.2.html), 2011 Feb 20, support ended 2016 Feb 20.
 * `concurrent.futures` module
 * introduction of `__pycache__` and versioned `.pyc` files
 * `math` module enhancements
 * `sys.get/setcheckinterval` deprecated in favour of `sys.get/setswitchinterval`

* [2.7](https://docs.python.org/3/whatsnew/2.7.html), 2010 Jul 3, support ended 2020 Jan 1.
 * end of the line for 2.*x*
 * some 3.1 features backported

* [3.1](https://docs.python.org/3/whatsnew/3.1.html), 2009 Jun 27, support ended 2012 June.
 * mostly fairly minor changes
 * `str.format` substitutions can now omit explicit indexes

* [3.0](https://docs.python.org/3/whatsnew/3.0.html), 2008 Dec 3, support ended 2009 Feb 13.
 * `print` is a function
 * strings are unicode; new `bytes` type for byte strings
 * type annotations
 * `nonlocal` statement
 * no more backticks (use `repr()` instead)
 * removed “<>” (use “!=” instead)
 * `exec` is a function, not a keyword
 * “/” now always does float division
 * all classes are “new-style”
 * default source encoding is now UTF-8

* [2.6](https://docs.python.org/3/whatsnew/2.6.html), 2008 Oct 1
 * preparation for 3.0
 * `with`-statement now always enabled
 * `multiprocessing` module
 * `str.format()` string formatting
 * `from __future__ import print_function`
 * allow Python-3-style `try` ... `except` *exc* `as` *name*
 * `from __future__ import unicode_literals`
 * `io` module
 * class decorators
 * `fractions` module
 * `future_builtins` module

* [2.5](https://docs.python.org/3/whatsnew/2.5.html), 2006 Sep 19
 * conditional expressions
 * `functools.partial`
 * absolute and relative imports
 * unified `try`/`except`/`finally`
 * `yield` is now an expression
 * `with`-statement, context managers enabled via `from __future__ import with_statement`
 * exceptions can now be new-style classes
 * `__index__` special method
 * `any()` and `all()` builtins
 * default source encoding is now enforced as ASCII

* [2.4](https://docs.python.org/3/whatsnew/2.4.html), 2005 Mar 30
 * built-in `set` and `frozenset`
 * generator expressions
 * `string.Template`
 * function/method decorators
 * `reversed()` builtin
 * `subprocess` module
 * `Decimal` data type
 * parenthesized import lists

* [2.3](https://docs.python.org/3/whatsnew/2.3.html), 2003 Jul 29
 * generator functions now always enabled
 * source code encodings (allowing for source to be written in e.g. UTF-8)
 * universal newlines
 * `enumerate()` and `sum()` builtins
 * `bool` type
 * extended slices
 * `assert` always asserts
 * `datetime` module enhancements

* [2.2](https://docs.python.org/3/whatsnew/2.2.html), 2001 Dec 21
 * “new-style” classes with descriptors
 * iterables &amp; iterators
 * generator functions: enable with `from __future__ import generators`
 * “//” integer division operator and `from __future__ import division`
 * nested function scopes now always enabled

* [2.1](https://docs.python.org/3/whatsnew/2.1.html), 2001 Apr 17
 * introduction of `from __future__ import` *whatever*
 * nested function scopes: enable with `from __future__ import nested_scopes`
 * rich comparison methods
 * warnings
 * weakrefs
 * arbitrary function attributes

* [2.0](https://docs.python.org/3/whatsnew/2.0.html), 2000 Oct 16
 * unicode strings
 * list comprehensions
 * string methods
 * garbage collection as a fallback from reference-counting
